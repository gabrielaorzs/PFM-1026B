```python
# lista de columnas con valores faltantes y su cantidad
cols_with_missing = [col for col in df.columns if df[col].isnull().any()]
for i in cols_with_missing:
    print(i,df[i].isnull().sum())
```

```python
# Lista de columnas a procesar
columns_to_process = ['reserve_id', 'hotel_id', 'customer_id']

# Lista de columnas a procesar
columns_to_process = ['reserve_id', 'hotel_id', 'customer_id']

# Iterar sobre las columnas y aplicar la imputación o eliminación de filas según corresponda
for column_name in columns_to_process:
    # Antes del cambio
    print(f"Valores faltantes en '{column_name}' antes del cambio:")
    print(df[column_name].isnull().sum())

    # Imputar valores faltantes o eliminar filas según la columna
    if column_name in ['reserve_id', 'hotel_id', 'customer_id']:
        # Eliminar filas con valores faltantes en 'reserve_id', 'hotel_id' o 'customer_id'
        df.dropna(subset=[column_name], inplace=True)
    else:
        # Imputar valores faltantes con el valor más frecuente
        most_frequent_value = df[column_name].mode()[0]
        df[column_name].fillna(most_frequent_value, inplace=True)

    # Después del cambio
    print(f"\nValores faltantes en '{column_name}' después del cambio:")
    print(df[column_name].isnull().sum())


```

```python
# Antes del cambio
print("Valores faltantes en 'checkin_date' antes del cambio:")
print(df['checkin_date'].isnull().sum())
```

# Nueva sección

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt


# Convertir las columnas de fecha al tipo datetime y date
df['reserve_datetime'] = pd.to_datetime(df['reserve_datetime'], errors='coerce')
df['checkin_date'] = pd.to_datetime(df['checkin_date'], errors='coerce').dt.date
df['checkout_date'] = pd.to_datetime(df['checkout_date'], errors='coerce').dt.date

# Definir la fecha de referencia
today = datetime.date.today()
```

```python
# Identificar filas con fechas futuras
future_reserve_dates = df[df['reserve_datetime'] > pd.Timestamp(today)]
future_checkin_dates = df[df['checkin_date'] > today]
future_checkout_dates = df[df['checkout_date'] > today]

# Contar filas con fechas futuras
num_future_reserve_dates = future_reserve_dates.shape[0]
num_future_checkin_dates = future_checkin_dates.shape[0]
num_future_checkout_dates = future_checkout_dates.shape[0]

print(f"Número de filas con fechas futuras en 'reserve_datetime': {num_future_reserve_dates}")
print(f"Número de filas con fechas futuras en 'checkin_date': {num_future_checkin_dates}")
print(f"Número de filas con fechas futuras en 'checkout_date': {num_future_checkout_dates}")
```

```python
print("Filas con fechas futuras en 'reserve_datetime':")
print(future_reserve_dates)

print("\nFilas con fechas futuras en 'checkin_date':")
print(future_checkin_dates)

print("\nFilas con fechas futuras en 'checkout_date':")
print(future_checkout_dates)
```

```python
# Eliminar filas con fechas futuras
df_cleaned = df[~((df['reserve_datetime'] > pd.Timestamp(today)) |
                  (df['checkin_date'] > today) |
                  (df['checkout_date'] > today))]
```

```python
# Validar que no hay fechas futuras
print("Número de fechas futuras en 'reserve_datetime':", df_cleaned[df_cleaned['reserve_datetime'] > pd.Timestamp(today)].shape[0])
print("Número de fechas futuras en 'checkin_date':", df_cleaned[df_cleaned['checkin_date'] > today].shape[0])
print("Número de fechas futuras en 'checkout_date':", df_cleaned[df_cleaned['checkout_date'] > today].shape[0])
```

```python
# Guardar los datos limpios
df_cleaned.to_csv('/content/reserve_cleaned.csv', index=False)
```
